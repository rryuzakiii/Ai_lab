```python
import numpy as np
```

```python
class LinearRegression:
    def __init__(self, learning_rate):
        self.m = 0
        self.b = 0
        self.alpha = learning_rate

    def mse(self, ypred, yreal):
       return (1 / len(yreal)) * np.sum((ypred - yreal)**2)

    def predict(self, xs):
        return self.m * xs + self.b
    
    def partials(self, xs, ys):
        dm = (2 / len(xs)) * np.sum((self.predict(xs) - ys) * xs)
        db = (2 / len(xs)) * np.sum((self.predict(xs) - ys) * 1)
        return dm, db
    
    def fit(self, xs, ys):
        for i in range(3000):

            ypred = self.predict(xs)
            mse = self.mse(ypred, ys)

            print(mse)

            dm, db = self.partials(xs, ys)
            self.m -= self.alpha * dm
            self.b -= self.alpha * db
```

```python
xs = np.array([1, 2, 3, 4, 6, 7])
ys = 3*xs + 5
```

```python
model = LinearRegression(learning_rate=0.006)
```

```python
model.fit(xs, ys)
```
